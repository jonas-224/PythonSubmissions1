## Section 1: Boolean Values — True and False

A **Boolean** is a data type with only two possible values: `True` or `False`. That's it — nothing else.

You've already been using Booleans without thinking about it. Every time you write an `if` statement, the condition inside it evaluates to either `True` or `False`:

```python
if age >= 16:    # age >= 16 is either True or False
```

Comparisons like `>`, `<`, `==`, `!=`, `>=`, `<=` all produce Boolean values. You can also store Booleans directly in variables.

````python
# EXAMPLE: Boolean values come from comparisons

# Each of these comparisons produces True or False
print(5 > 3)        # True — 5 is greater than 3
print(5 == 3)       # False — 5 does not equal 3 (== is "equals", = is assignment)
print(10 <= 10)     # True — 10 is less than or equal to 10
print("a" != "b")   # True — "a" is not equal to "b"

# You can store a Boolean in a variable, just like a number or string
is_raining = True
has_umbrella = False

# Check the type — it's 'bool'
print(type(is_raining))   # <class 'bool'>

# You can use Boolean variables directly in if statements
if is_raining:
    print("Grab an umbrella!")
else:
    print("Enjoy the sunshine!")
````

## Section 2: The AND Operator

The `and` operator combines two Boolean values. The result is `True` **only when BOTH sides are True**. If either side is False, the whole thing is False.

**Real-world example:** "I'll go to the beach if it's warm AND sunny." Both conditions must be true — warm but rainy? Nope. Sunny but cold? Nope. Only warm AND sunny works.

### The AND Truth Table

A **truth table** shows every possible combination of inputs and what the output is:

| A | B | A and B |
|---|---|---------|
| True | True | **True** |
| True | False | **False** |
| False | True | **False** |
| False | False | **False** |

**Memory trick:** AND is strict — it only says True when *everything* is True.

````python
# EXAMPLE: The AND operator in action

# All four possible combinations
print(f"True  and True  = {True and True}")    # True — both are True
print(f"True  and False = {True and False}")   # False — second is False
print(f"False and True  = {False and True}")   # False — first is False
print(f"False and False = {False and False}")  # False — both are False

# A practical example: can you drive?
has_license = True
has_car = False

# You need BOTH a license AND a car to drive
can_drive = has_license and has_car   # True and False = False
print(f"\nHas license: {has_license}")
print(f"Has car: {has_car}")
print(f"Can drive? {can_drive}")       # False — missing the car
````

## Section 3: The OR Operator

The `or` operator combines two Boolean values. The result is `True` **when at least one side is True**. It's only False when *both* sides are False.

**Real-world example:** "I'll be happy if I get pizza OR tacos." Pizza? Happy. Tacos? Happy. Both? Happy. Neither? Sad.

### The OR Truth Table

| A | B | A or B |
|---|---|--------|
| True | True | **True** |
| True | False | **True** |
| False | True | **True** |
| False | False | **False** |

**Memory trick:** OR is generous — it says True as long as *at least one* input is True.

````python
# EXAMPLE: The OR operator in action

# All four possible combinations
print(f"True  or True  = {True or True}")    # True — at least one is True
print(f"True  or False = {True or False}")   # True — first one is True
print(f"False or True  = {False or True}")   # True — second one is True
print(f"False or False = {False or False}")  # False — NEITHER is True

# A practical example: can you get into the movie?
is_18_or_older = False
has_parent_with_them = True

# You can get in if you're 18+ OR if you have a parent with you
can_enter = is_18_or_older or has_parent_with_them  # False or True = True
print(f"\nIs 18+: {is_18_or_older}")
print(f"Has parent: {has_parent_with_them}")
print(f"Can enter? {can_enter}")       # True — parent is with them
````

## Section 4: The NOT Operator

The `not` operator is the simplest — it just **flips** the value. True becomes False. False becomes True.

| A | not A |
|---|-------|
| True | **False** |
| False | **True** |

**Real-world example:** "If it's NOT raining, let's go outside." `not True` = False (it IS raining, so we stay in). `not False` = True (it's NOT raining, so we go out).

````python
# EXAMPLE: The NOT operator flips True/False

print(f"not True  = {not True}")    # False — flipped from True
print(f"not False = {not False}")   # True — flipped from False

# Practical example
is_raining = True

# We want to go outside if it's NOT raining
if not is_raining:                   # not True = False, so this block is skipped
    print("Let's go outside!")
else:
    print("Better stay in.")

# Change the weather
is_raining = False
if not is_raining:                   # not False = True, so this block runs
    print("Now let's go outside!")
````

## Quick Reference

| Operator | True When | Memory Trick |
|----------|-----------|-------------|
| `A and B` | Both are True | "Both must pass" |
| `A or B` | At least one is True | "Either works" |
| `not A` | A is False (flips it) | "Flip it" |

**Reminder that there is an order of operations:** `not` first, then `and`, then `or`. Use parentheses to be safe!

**Exam tip:** When you see a compound condition, evaluate it step by step — do `not` first, then `and`, then `or`. Write out each step. Don't try to do it all in your head.
